```python
from matplotlib import pyplot as plt
from Geo2d import *
from symetries import *
```

```python
fig, ax = plt.subplots()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
P= Point(1,1,'P')

d1 = Droite(1,2,3)
d2 = P.symetrie_centrale(d1)
P.draw(ax)
d1.draw(ax)
d2.draw(ax)

A1  = Point(2,1,'A1')
A2  = P.symetrie_centrale(A1)
A2.nom=('A2')
A1.draw(ax,dx=0.2)
A2.draw(ax,dx=-0.7)
fig.suptitle('Symétries centrales de centre P')
```

```python
d  = Droite(1,2,3)
d1 = Droite(2,1,3)
d2 = retournement(d1,d)
fig, ax = plt.subplots()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
d.draw(ax,label="d")
d1.draw(ax,label="d1")
d2.draw(ax,label="d2")
ax.legend()
fig.suptitle("d2 est l'image de d1 par le retounement d'axe d")
```

```python
d  = Droite(1,2,3)
d1 = Droite(2,1,3)
d2 = retournement(d1,d)
fig, ax = plt.subplots()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
d.draw(ax,label='d')
d1.draw(ax,label='d1')
d2.draw(ax,label='d2')
fig.suptitle("d2 est l'image de d1 par le retounement d'axe d")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
P = Point(2,2,'P')
P.draw(ax,0.2)
d = Droite(1,2,3)
H = projection(P,d, nom='H')
PH = Segment(P,H)
Q = retournement(P,d,'Q')
R = d.retournement(P)
R.nom='R'
PH.draw(ax,lw=1)
d.draw(ax,label='d')
H.draw(ax,0.2)
ax.set_aspect('equal')

Q.draw(ax,dx=0.1)
R.draw(ax,dx=-0.44)
ax.legend()
fig.suptitle("H est la projection orhogonale de P sur d")
```
