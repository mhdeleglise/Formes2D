```python
from matplotlib import pyplot as plt
from Geo2d import *
```

```python
fig, ax = plt.subplots()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
P= Point(1,1,'P')

d1 = Droite(1,2,3)
d2 = P.symetrie_centrale(d1)
P.draw(ax)
d1.draw(ax)
d2.draw(ax)

A1  = Point(2,1,'A1')
A2  = P.symetrie_centrale(A1)
A2.nom=('A2')
A1.draw(ax,dx=0.2)
A2.draw(ax,dx=-0.7)
fig.suptitle('Symétries centrales de centre P')
```

```python
d  = Droite(1,2,3)
d1 = Droite(2,1,3)
d2 = retournement(d1,d)
fig, ax = plt.subplots()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
d.draw(ax,label="d")
d1.draw(ax,label="d1")
d2.draw(ax,label="d2")
ax.legend()
fig.suptitle("d2 est l'image de d1 par le retounement d'axe d")
```

```python
fig, ax = plt.subplots()
ax.set_xlim(-8,8)
ax.set_ylim(-6,5)
P = Point(2,2,'P')
P.draw(ax,0.2)
d = Droite(1,2,3)
H = projection(P,d, nom='H')
PH = Segment(P,H)
PH.draw(ax,lw=1)
d.draw(ax,label='d')
H.draw(ax,0.2)
ax.set_aspect('equal')
ax.legend()
fig.suptitle("H est la projection orhogonale de P sur d")
```

```python
fig, ax = plt.subplots(figsize = (7,7))
ax.set_xlim(-2,3)
ax.set_ylim(-2,3)
d = Droite(2,1,3)
P = Point(2,2,'P')
d1= parallele(d,P)
d.draw(ax,label='d')
d1.draw(ax,label= 'd1')
P.draw(ax)
fig.suptitle("d1, parallèle à d passant par P", fontsize=16, color='green')
ax.set_aspect('equal')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.set_xlim(-3.0,5.0)
ax.set_ylim(-4.0,4.0)
d = Droite(1,2,3)
P = Point(2,2,'P')
d1= orthogonale(d,P)
d.draw(ax,label='d')
d1.draw(ax,label='perpendiculaire')
P.draw(ax,dx= -0.7)
#plt.margins(0,0)
ax.legend()
ax.set_aspect('equal')
fig.suptitle("Perpendiculaire à d passant par P")
fig.set_facecolor('yellow')
print(ax)
```
