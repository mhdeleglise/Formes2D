```python
d1 = Droite(1,2,3);
d2 = Droite(2,1,3)
print(d1,d2)
P = d1.intersection(d2);
fig, ax = plt.subplots()
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
P.draw(ax,s=30)
d1.draw(ax)
d2.draw(ax)
fig.suptitle("Intersection de droites")
ax.set_aspect('equal')
```

```python
d  = Droite(1,2,3)
d1 = Droite(2,1,3)
fig, ax = plt.subplots()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')
d2 = retournement(d1,d)
d.draw(ax,label="d")
d1.draw(ax,label="d1")
d2.draw(ax,label="d2")
plt.title("d2 retournement de d1 autour de d", color='green')
ax.legend()
```

```python
d1 = Droite(1,2,3);
d2 = Droite(2,1,3)
d = d1.bissectrice(d2)
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
d1.draw(ax,label='d1',color='blue')
d2.draw(ax,label='d2',color='orange')
d.draw(ax,label='d',color='green')
ax.legend()
fig.suptitle("Bissectrice")
print(degrees(d1.angle(d2)))
print(degrees(d1.angle(d)))
```
