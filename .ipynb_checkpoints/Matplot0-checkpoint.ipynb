```python
X = [3*k/100 for k in range(10)]
print("X = ",X)
print("Type de X = ", type(X))
Y = [x**2 for x in X]
Z = [x**3 for x in X]
plt.plot(X,Y)
plt.plot(X,Z)
```

```python
Z = [x**3 for x in X]
l1 = plt.plot(X,Y, label="y=x^2")
l2 = plt.plot(X,Z, label="y=x^3")
plt.title("Courbes y = x^2, y = x^3",color="green")
plt.legend()
plt.xlabel(" Les abscisses x ", color = 'blue')
plt.ylabel(" Les ordonnées y ", color='blue')
```

```python
""" En utilisant numpy """
import numpy as np
```

```python
""" On retrouve notre tableau des valeurs de x de 0 à 3, espacées de 0.03 """
x = np.arange(0,3,3/100); print(x)
print("len(x) = ", len(x), "   type(x): ", type(x))
```

""" 
Remarquons que x n'est plus une liste de réels c'est un tableau numpy. Les acces aux 
tableauxsont sont plus rapides que les accès aux éléments d'une liste. Pour plus
de détail
https://numpy.org/doc/stable/reference/generated/numpy.ndarray.htmls:
"""

```python
""" Si x est un tableau de nombres créé par np.arange, par exemple, la fonction np.square 
(resp. np.cos()) renvoie le tableau des carrés (res. des cosinus) des éléments de x """
plt.plot(x,np.square(x), color = 'blue', label='x^2')
plt.plot(x,np.cos(x), color = 'orange', label= 'cos(x)')
plt.grid(True)
plt.legend()
```

```python
plt.axes().set_aspect('equal') # Meme echelle sur les deux axes
plt.plot(x,np.square(x), color = 'blue', label='x^2')
plt.plot(x,np.cos(x), color = 'orange', label= 'cos(x)')
plt.grid(True)
plt.legend()
```

```python
X  = [1,2,3,4,5]
Y  = [1.5,4,5,3.4,1]
Y2 = [1.,3.7,2,2,0.9] 
Y3 = [1,2, 3, 3, 2]
X1 = [1.0, 3.5]
Y1 = [2.5, 3.0]

plt.plot(X, Y,  '-ro', ms='8',lw=1) # ligne continue liant desronds rouges
plt.plot(X, Y2, '--go', ms ='2') # tirets reliant des ronds verts
plt.plot(X1,Y1, 'co-', lw=4) # couleur cyan, ronds reliés par des traits de largeur 4
plt.plot(X,Y3,'o')  # ronds, couleur par défaut
```

```python
X  = [1,2,3,4,5]
Y  = [1.5,4,5,3.4,1]
Y2 = [1.,3.7,2,2,0.9] 
X1 = [1.0, 3.5]
Y1 = [2.5, 3.0]

plt.plot(X,Y,'--r', marker='s',ms=8) # marker est un carré
plt.plot(X,Y2,':bs') # pointillés reliant des carrés bleus
```

```python
x = np.linspace(0,1,40)
y = x*x
plt.plot(x,y,'-ro') # points reliés par un trait continu, en rouge
plt.plot(x,y/2,'-g') # points non marqués reliés par un trait continu vert
plt.plot(x,y/3,'s',ms=4) # points marqués par des carrés de taille 4, couleur par défaut 
```
