```python
import matplotlib.pyplot as plt
from Geo2d import *
from symetries import *
```

```python
d1 = Droite(1,2,3);
d2 = Droite(2,1,3)
print(d1,d2)
P = d1.intersection(d2);
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
P.draw(ax,s=30)
d1.draw(ax)
d2.draw(ax)
fig.suptitle("Intersection de droites")
```
