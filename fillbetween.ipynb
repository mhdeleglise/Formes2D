```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,100)
```

```python
y1 = np.sqrt(1-x*x)
y2 = -y1
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.axis('equal')
plt.fill_between(x,y1,y2)
```

```python
def draw_circle(x0, y0, r):
    x  = np.linspace(x0-r, x0 + r,100)
    y1 = y0 + np.sqrt(r**2 - (x-x0)*(x-x0))
    y2 = y0 - np.sqrt(r**2 - (x-x0)*(x-x0))
    plt.axis('equal')
    plt.grid(True)
    plt.fill_between(x,y1,y2, color='green')
```

```python
draw_circle(2,2,1)
```
