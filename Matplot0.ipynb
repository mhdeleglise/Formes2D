```python
X = [3*k/100 for k in range(100)]
print(X)
print("Type de X = ", type(X))
Y = [x**2 for x in X]
Z = [x**3 for x in X]
plt.plot(X,Y)
plt.plot(X,Z)
```

```python
Z = [x**3 for x in X]
plt.plot(X,Y, label="y=x^2")
plt.plot(X,Z, label="y=x^3")
plt.title("Courbes y = x^2, y = x^3",color="green")
plt.legend()
plt.xlabel(" Les abscisses x ", color = 'blue')
plt.ylabel(" Les ordonnées y ", color='blue')
```

```python
""" En utilisant numpy """
import numpy as np
```

```python
""" On retrouve notre tableau des valeurs de x de 0 à 3, espacées de 0.03 """
x = np.arange(0,3,3/100); print(x)
print("len(x) = ", len(x), "   type(x): ", type(x))
```

```python
""" 
Remarquons que x n'est plus une liste de réels c'est un tableau numpy. Les acces aux 
tableauxsont sont plus rapides que les accès aux éléments d'une liste. Pour plus
de détail
https://numpy.org/doc/stable/reference/generated/numpy.ndarray.htmls:
"""
```

```python
""" Si x est un tableau de nombres créé par np.arange, par exemple, la fonction np.square 
(resp. np.cos()) renvoie le tableau des carrés (res. des cosinus) des éléments de x """
plt.plot(x,np.square(x), color = 'blue', label='x^2')
plt.plot(x,np.cos(x), color = 'orange', label= 'cos(x)')
plt.grid(True)
plt.legend()
```

```python
""" Si x est un tableau de nombres créé par np.arange, par exemple, la fonction np.square 
(resp. np.cos()) renvoie le tableau des carrés (res. des cosinus) des éléments de x """
plt.axes().set_aspect('equal') # Meme echelle sur les deux axes
plt.plot(x,np.square(x), color = 'blue', label='x^2')
plt.plot(x,np.cos(x), color = 'orange', label= 'cos(x)')
plt.grid(True)
plt.legend()
```
