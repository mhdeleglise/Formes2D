```python
from Polygones import *
from intersections import *
```

```python
A, B, C, D, E = Point(0,0,'A'), Point(1,1,'B'), Point(0,1,'C'), Point(1,0,'D'), Point(0.5, 1.5, 'E')
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
for p in [A,B,C,D,E]: p.draw(ax)
q = Polygone([A,B,C,D,E])
q.draw(ax,color='orange')
pmq=q.multiplicities()
for p in pmq:
    p.draw(ax,color='red',s=40)
pmq
print("Aire(q)= ", q.aire())
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
for p in [B,C,D,E]: p.draw(ax)
q = Polygone([B,C,D,E])
q.draw(ax,color="green")
pmq = q.multiplicities()
for p in pmq:
    p.draw(ax,color='red',s=40)
pmq
I = pmq[0]
I.nom=('I')
I.draw(ax)
print("Aire(q)= ", q.aire())
print("aire(Triangle(C,D,I)= ",Triangle(C,D,I).aire())
print("aire(Triangle(B,E,I)= ",Triangle(B,E,I).aire())
```

```python
q.is_simple()
```

```python
A, B, C = Point(0,0), Point(0.5,0.5), Point(1,0)
```

```python
F, E, D = [p + Vecteur(0,1) for p in [A,B,C]]
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
for p in [A,B,C,D,E,F]:
    p.draw(ax)
q = Polygone([A,B,C,D,E,F])
q.draw(ax,color='orange')
```
