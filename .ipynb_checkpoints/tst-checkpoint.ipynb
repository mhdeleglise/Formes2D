```python
A = Point(0, 1,'A')
B = Point(2, -2,'B')
C = Point(-1, -1,'C')
```

```python
fig,ax = plt.subplots()
ax.set_xlim(-1.3,2.5)
ax.set_ylim(-2.3,1.5)
ABC = Polygone([A,B,C])
A.draw(ax,namecolor = 'red', dx=0.08, s=100)
B.draw(ax,namecolor = 'blue',dx = 0.05, s= 50)
C.draw(ax,namecolor = 'black', dx= -0.15,s=1)
ABC.draw(ax, color='green', namecolor='red', ms=0)
print(type(ABC), "Aire = ABC.aire()")
```

```python
AB=Segment(A, B)
AC=Segment(A,C)
BC=Segment(B,C)
ABC=Polygone([A,B,C])
print("Aire de ABC ", ABC.aire())
G = (A+B+C)/3
G.nom="G"
print(A,B,C,G)

fig, ax = plt.subplots()
A.draw(ax,  color='red', s=200,dx=0.1)
B.draw(ax,  color='green',dx=0.05)
C.draw(ax, s=40, dx = 0.06)
AB.draw(ax, color='red')
BC.draw(ax, color='blue')
AC.draw(ax, color='green')
G.draw(ax,color='orange', namecolor='black', s=100, label="Centre de gravité")
ax.legend()
```

```python
A = Point(-2,0)
B = Point(2,0)
C = Point(0.5,1.5)
ABC = Triangle(A,B,C)
I = ABC.centreCercleInscrit('I')
c = ABC.cercleInscrit()
ce = Cercle.cercle((A+B)/2, (B+C)/2, (C+A)/2)

# Tracés
fig, ax = plt.subplots()
ax.set_aspect(1)
ABC.draw(ax)
I.draw(ax,dy=0.05)
ax.axis('off')
c.draw(ax,color='red',label="Cercle Inscrit")
ce.draw(ax,color='blue',label="Cercle d'Euler")
ax.legend()
```

```python
A = Point(0.0,0,'A')
B = Point(2,0,'B')
C = Point(1.0,1.0,'C')
ABC = Triangle(A,B,C)
AB = droite(A,B)
AC = droite(A,C)
BC = droite(B,C)
d0 = ABC.bissectrice(0)
d1 = ABC.bissectrice(1)
d2 = ABC.bissectrice(2)
c0 = ABC.cercleExInscrit(0)
c1 = ABC.cercleExInscrit(1)
c2 = ABC.cercleExInscrit(2)
c  = ABC.cercleInscrit()
d00 = d0.orthogonale(A)
d11 = d1.orthogonale(B)
d22 = d2.orthogonale(C)


fig, ax = plt.subplots(constrained_layout=True,figsize=(10,10))
ax.set_aspect(1)
ax.set_xlim(-2.7,4)
ax.set_ylim(-5,2.3)
A.draw(ax,dx=-0.15,dy=-0.06,namecolor='orange')
B.draw(ax,dx=0.10,dy=-0.08,namecolor='orange')
C.draw(ax,dy=0.08,dx=-0.07,namecolor='orange')
ABC.draw(ax,color='cyan',lw=2)

d0.draw(ax,color='green')
d1.draw(ax,color='green')
d2.draw(ax,color='green')

AC.draw(ax,color='black')
AC.draw(ax,color='black')
BC.draw(ax,color='black')

d00.draw(ax,color='violet')
d22.draw(ax,color='violet')
d11.draw(ax,color='violet')

c.draw(ax,color='red')
c0.draw(ax, color='green')
c1.draw(ax, color='green')
c2.draw(ax, color='green')

fig.suptitle('Cercles inscrit et  exinscrits',fontsize=20)
ax.axis('off')
```
