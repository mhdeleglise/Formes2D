```python
A = Point(0,0,'A')
B = Point(4.5,0,'B')
C = Point(1.5,4,'C')
I = milieu(B,C)
J = milieu(C,A)
K = milieu(A,B)
ABC = Triangle(A,B,C)
G   = ABC.barycentre('G')
A0  = ABC.piedHauteur(0) # pied de la hauteur issue de A = sommet(0)
B0  = ABC.piedHauteur(1)
C0  = ABC.piedHauteur(2)
h0  = ABC.hauteur(0,limited = True) 
h1  = ABC.hauteur(1,limited = True)
h2  = ABC.hauteur(2,limited = True)
cercle_euler = Triangle(A0,B0,C0).cercleCirconscrit() 
O  = ABC.centreCercleCirconscrit('O')
H  = ABC.orthocentre(nom='H')
a  = milieu(H,A)
b  = milieu(H,B)
c  = milieu(H,C)

""" Tracé de l'image """
fig,ax = plt.subplots(figsize=(10, 10.5))
#plt.subplots_adjust(bottom=0.1, right=0.8, top=0.85)

ABC.draw(ax,color = 'cyan',lw=0.)
A.draw(ax,s=100,color='blue',dx=-0.2)
B.draw(ax,s=100,color='blue',dx= 0.1)
C.draw(ax,s=100,color='blue',dx= 0.1)
cercle_euler.draw(ax,color='crimson',facecolor='floralwhite',alpha=0.5, lw=1)

mediatrice(A,B).draw(ax,color='blue',label='médiatrices',lw=1)
mediatrice(B,C).draw(ax,color='blue',lw=1)
mediatrice(C,A).draw(ax,color='blue',lw=1)
h0.draw(ax,color='green',label="hauteurs")
h1.draw(ax,color='green')
h2.draw(ax,color='green')
Segment(A,I).draw(ax,color='darkorange',label='medianes',lw=1.2)
Segment(B,J).draw(ax,color='darkorange',lw=1.2)
Segment(C,K).draw(ax,color='darkorange',lw=1.2)

O.draw(ax,label="O Centre cercle circonscrit",dx=0.1,color='blue',namecolor='blue',s=40)
H.draw(ax,label="H Orthocentre",dx=-0.10,dy=0.08,color='green',namecolor='green',s=40)
G.draw(ax,label="G Centre de gravité",dy=0.05,color='darkorange',namecolor='orange',s=40)

for P in [I,J,K,A0,B0,C0,a,b,c]:
    P.draw(ax,dx = 0.10, dy=0.01,color='black',namecolor='blue',s=30)
fig.suptitle("Cercle d'Euler, ou Cercle des 9 points ", fontsize = 24, color = 'green')
ax.text(-0.10,4.5,"Le cercle d'Euler du triangle ABC, d'orthocentre H, passe par les 9 points noirs:\n\
les 3 milieux des côtés, les 3 pieds des hauteurs et les mileux de AH, BH et CH", fontsize = 13, weight='light',color='darkorange')
ax.set_aspect('equal')
ax.legend()
ax.axis('off')
fig.savefig('euler.png')      
```

```python
A = Point(0,0,'A')
B = Point(6,0,'B')
C = Point(0.5,4,'C')
I = milieu(B,C)
J = milieu(C,A)
K = milieu(A,B)
ABC = Triangle(A,B,C)
G   = ABC.barycentre('G')
A0  = ABC.piedHauteur(0) # pied de la hauteur issue de A = sommet(0)
B0  = ABC.piedHauteur(1)
C0  = ABC.piedHauteur(2)
h0  = ABC.hauteur(0,limited = True) 
h1  = ABC.hauteur(1,limited = True)
h2  = ABC.hauteur(2,limited = True)
cercle_euler = Triangle(A0,B0,C0).cercleCirconscrit() 
O  = ABC.centreCercleCirconscrit('O')
H  = ABC.orthocentre(nom='H')
a  = milieu(H,A)
b  = milieu(H,B)
c  = milieu(H,C)

b0 = ABC.bissectrice(0)
b1 = ABC.bissectrice(1)
b2 = ABC.bissectrice(2)


""" Tracé de l'image """
fig,ax = plt.subplots(figsize=(10, 9.5))

ABC.draw(ax,color = 'cyan',lw=0.5)
A.draw(ax,s=100,color='blue',dx=-0.2)
B.draw(ax,s=100,color='blue',dx= 0.1)
C.draw(ax,s=100,color='blue',dx= 0.1)
cercle_euler.draw(ax,color='crimson',alpha=0.5, lw=1)

mediatrice(A,B).draw(ax,color='blue',label='médiatrices',lw=1)
mediatrice(B,C).draw(ax,color='blue',lw=1)
mediatrice(C,A).draw(ax,color='blue',lw=1)
h0.draw(ax,color='green',label="hauteurs")
h1.draw(ax,color='green')
h2.draw(ax,color='green')
Segment(A,I).draw(ax,color='darkorange',label='medianes',lw=1.2)
Segment(B,J).draw(ax,color='darkorange',lw=1.2)
Segment(C,K).draw(ax,color='darkorange',lw=1.2)

Q = ABC.centreCercleInscrit('Q')
O.draw(ax,label="O Centre cercle circonscrit",dx=0.1,color='blue',namecolor='blue',s=40)
H.draw(ax,label="H Orthocentre",dx=-0.10,dy=0.08,color='green',namecolor='green',s=40)
G.draw(ax,label="G Centre de gravité",dy=0.05,color='darkorange',namecolor='orange',s=40)
b0.draw(ax,color='fuchsia',lw=0.5)
b1.draw(ax,color='fuchsia',lw=0.5)
b2.draw(ax,color='fuchsia',lw=0.5)



for P in [I,J,K,A0,B0,C0,a,b,c]:
    P.draw(ax,dx = 0.10, dy=0.01,color='black',namecolor='blue',s=30)
fig.suptitle("Cercle d'Euler et Cercle Inscrit", fontsize = 24, color = 'green')
ax.set_aspect('equal')
ci = ABC.cercleInscrit()
Q.draw(ax,label="Q Centre cercle inscrit",color = 'fuchsia', s=40)
cercle_euler.centre.draw(ax,label="euler.centre")
ci.draw(ax)
ax.legend()
ax.axis('off')
fig.savefig('euler.png')      
```
