```python
fig,ax = plt.subplots(figsize=(10,4))
A=Point(-2,0,'A')
B=Point(+2,0,'B')
C=Point(0,1,'C')
A.draw(ax,dx=-0.15)
B.draw(ax,dx=0.05)
ABC=Triangle(A,B,C)
ABC.draw(ax)
P = ABC.centreCercleInscrit('Q')
c = ABC.cercleInscrit()
P.draw(ax,dy= 0.05,label="Centre cercle")
c.draw(ax)
d1 = ABC.bissectrice(0)
d2 = ABC.bissectrice(1)
d1.draw(ax,label='d1')
d2.draw(ax,label='d2')
ax.set_aspect('equal')
ax.legend()
ax.axis('off')

```
