```python
" plt.subplots() crée une figure qui contient plusieurs objets de type Axes "
" Un objet de type Axes  est une plage rectangulaire sur laquelle "
" On peut tracer des points, des lignes, .... "
fig, ax = plt.subplots() # ax est un objet de type Axes (c'est une portion d'espace de limites par défaut 0 1 0 1
print("type fig = ", type(fig))
print("type ax = ", type(ax))
print("Limites en x: ", ax.get_xlim()) # Les limites en x 
print("Limites en y: ", ax.get_ylim()) # Les limites en y 
print(ax)
fig.set_facecolor('yellow')
```

```python
xdata  = np.arange(0,1.1,0.01)
ydata1 = np.sqrt(xdata)
ydata2 = np.cos(xdata)
```

```python
fig, (ax1, ax2) = plt.subplots(2) # 2 lignes de graphes 
fig.suptitle('2 Axes superposés')
ax1.plot(xdata,ydata1)
ax2.plot(xdata,ydata2,color='orange')
ax1.grid(True)
ax2.grid(True)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2) # 1 ligne de 2 graphes 
fig.suptitle('2 Axes côte à côte qui ne partagent pas leurs axes verticaux')
ydata1 = np.sqrt(xdata)
ydata2 = np.cos(xdata)
ax1.plot(xdata,ydata1)
ax2.plot(xdata,ydata2,color='orange')
ax1.grid(True)
ax2.grid(True)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2) # 1 ligne de 2 graphes 
fig.suptitle(' Les deux même avec les memes limites en y' )
ydata1 = np.sqrt(xdata)
ydata2 = np.cos(xdata)
ax1.plot(xdata,ydata1)
ax2.plot(xdata,ydata2,color='orange')
ax1.grid(True)
ax2.grid(True)
ax1.sharey(ax2) # Les mêmes limites en y pour les 2 Axes (cad. les 2 subplots)
```

```python
import matplotlib.pyplot as plt
import numpy as np

ydata0 = np.exp(-xdata)
ydata1 = np.cos(xdata)
ydata2 = np.sin(xdata)
ydata3 = np.sqrt(xdata)

fig, axs = plt.subplots(2, 2)
axs[0,0].plot(ydata0,color='orange',label='y = exp(-x)'); axs[0,0].legend()
axs[0,1].plot(ydata1,color='blue',label='y=cos(x)'); axs[0,1].legend()
axs[1,0].plot(ydata2,color='violet',label='y=sin(x)');axs[1,0].legend()
axs[1,1].plot(ydata3, label='y=sqrt(x');axs[1,1].legend()
fig.suptitle('4 Axes',fontsize=16, color='green')
```
