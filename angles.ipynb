```python
from Geo2d import *
from Polygones import *
```

```python
A=Point(-1,1,'A')
B=Point(3,1,'B')
C=Point(1,2,'C')
d1 = droite(A,B)
d2 = droite(A,C)
d3 = d1.bissectrice(d2)
d4 = d1.bissectrice(d2,1)

fig, ax = plt.subplots()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')
d1.draw(ax,color='blue')
d2.draw(ax,color='blue')
d3.draw(ax,color="green")
d4.draw(ax,color="orange")
fig.suptitle("Les deux bissectrices des droites bleues d1 et d2")
```

```python
fig, ax = plt.subplots()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')
d1 = demi_droite(A,B)
d2 = demi_droite(A,C)
d3 = d1.bissectrice(d2)
d1.draw(ax,color='blue')
d2.draw(ax,color='blue')
d3.draw(ax,color="green")
fig.suptitle("L'unique bissectrices des demi-droites d1 et d2")
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(-2,3)
ax.set_ylim(-1,2)
ax.grid(True)
A=Point(-1,0,'A')
B=Point(1,0,'B')
C=Point(0,1.5,'C')
ABC=Triangle(A,B,C)
A.draw(ax)
B.draw(ax)
C.draw(ax)
ABC.draw(ax,color='orange',ms=4)
AB = droite(A,B)
AC = droite(A,C)
vAB = vecteur(A,B)
vAC = vecteur(A,C)
ax.text(1.5,1.5,"Angle (droite(AB),droite(AC)) = {:.1f}".format(degrees(AB.angle(AC))))
ax.text(1.5,1,"Angle (droite(AC),droite(AB)) = {:.1f}".format(degrees(AC.angle(AB))))
ax.text(1.5,0.5,"Angle (vecteur(AB),vecteur(AC)) = {:.1f}".format(degrees(vAB.angle(vAC))))
ax.text(1.5,0.0,"Angle (vecteur(AC),vecteur(AB)) = {:.1f}".format(degrees(vAC.angle(vAB))))
ax.text(-1.8,-0.5," Les angles de vecteurs sont dfinis modulo 2pi, les angles de droites ne sont dfinis que modulo pi")
fig.suptitle("Les angles sont orients",y=0.8)
ax.axis('off')
```
