```python
from Polygones import *
```

```python
A = Point(0,0,'A')
B = Point(0,2,'B')
C = Point(2,4,'C')
ABC = Triangle(A,B,C)
G=ABC.barycentre('G')
```

```python
fig,ax= plt.subplots()
ax.set_xlim(-6,4)
ax.set_ylim(-1,5)
A.draw(ax)
B.draw(ax)
C.draw(ax)
G.draw(ax)
ABC.draw(ax)
hauteur(C,ABC).draw(ax)
hauteur(A,ABC).draw(ax)
hauteur(B,ABC).draw(ax)
ax.set_aspect('equal')
ax.axis('off')
```
